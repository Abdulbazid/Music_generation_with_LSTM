# Preprocessing all the files


```python
notes = []

for file in glob.glob(r"C:\Users\Abdul\Desktop\music\albeniz/*.mid"):
    midi = converter.parse(file)
    
    
    print("parsing %s" %file)
    
    elements_to_parse = midi.flat.notes
    
    for ele in elements_to_parse:
        # ifthe element is a note,       
        if isinstance(ele,note.Note):
            notes.append(str(ele.pitch))     ## isinstance a fuction which checks element is notes or chords
            
        # if the element isa chord
        elif isinstance(ele,chord.Chord):
            notes.append("+".join(str(n) for n in ele.normalOrder))
    
```

```python
notes
```

```python
len(notes)
```

```python
with open("notes" , "wb") as filepath:
    pickle.dump(notes, filepath)
```

```python
with open("notes" , "rb") as f:
    notes = pickle.load(f)
```

```python
n_vocab = len(set(notes))      ## number of differnt classes 
```

```python
print("total notes:", len(notes))
print("unique notes:",n_vocab)
      
```

# prepare Sequential Data for LSTM

```python
## how many elements lstm input should consider
sequence_length =100
```

```python
pitch_names = sorted(set(notes))
```

```python
## Mapping between element to int value

ele_to_int = dict( (ele, num) for num ,ele in enumerate(pitch_names))
```

```python
print(ele_to_int)
```

```python
network_input = []
network_output = []

```

```python
for i in range(len(notes) - sequence_length):
    seq_in = notes[i: i + sequence_length]  ## contains 100 values
    seq_out = notes[i + sequence_length]
    
    network_input.append([ele_to_int[ch] for ch in seq_in])
    network_output.append(ele_to_int[seq_out])
```

```python
## No. of examples

n_patterns = len(network_input)
print(n_patterns)
```

```python
network_input = np.reshape(network_input, (n_patterns , sequence_length,1))    ## 1 is for converting into 3 dimension beacuse LSTM take 3 dimension 
```

```python
network_input.shape
```

```python
normalise_network_input = network_input/float(n_vocab)
```

```python
##  Network output are teh classes , encode into onr hot vector
!pip install keras-utils
```

```python
from keras.utils import to_categorical
```

```python
network_output = to_categorical(network_output)
```

# create Model


```python
from keras.models import Sequential
from keras.layers import *
from keras.callbacks import ModelCheckpoint , EarlyStopping
```

```python
model = Sequential()

model.add( LSTM( units = 512,
               input_shape = (normalise_network_input.shape[1], normalise_network_input.shape[2]),
               return_sequences = True))

model.add( Dropout(0.3))

model.add( LSTM(512,return_sequences = True))
model.add(Dropout(0.3))
model.add( LSTM(512))
model.add(Dense(256))
model.add( Dropout(0.3))
model.add( Dense(n_vocab, activation = "softmax"))
```

```python
model.compile(loss = "categorical_crossentropy", optimizer = "adam")
```

```python
model.summary()
```

```python
checkpoint = ModelCheckpoint("model.hdf5", monitor ='loss', verbose = 0, save_best_only = True, mode = 'min')


model_his = model.fit(normalise_network_input, network_output, epochs = 100, batch_size = 128 , callbacks = [checkpoint])
```
